# 분류할 이미지 file 위치를 먼저 확인한다.

```python
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
```

```python
from matplotlib.pyplot import imshow
%matplotlib inline
# (%matplotlib inline)는 이미지를 이곳에 보여달라는 의미
```

# 파일 위치 확인은 !ls로 할수 있으나, 안되므로 dir로..
!dir d:\data\python\cnn_image

```python
train_data = torchvision.datasets.ImageFolder (root='d:\data\python\cnn_image\Origin_data', transform=None)
```

```python
# 이미지가 폴더로 구분되어 있으므로, 이를 라벨로 인식하여 전체 이미지파일을 분류할 수 있다.
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
```

```python
# 이미지 확인을 하는데, 이미지가 많으므로 'break'를 걸어서 첫장만 확인
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    imshow(data)
    break
```

# 이미지가 크므로, Resize를 호출해서 크기를 줄이고,한꺼번에 자동으로 적용하기 위해 Compose를 활용
# 그리고 rain_data에 이를 적용하기 위해 transform 조건에 연결

```python
trans = transforms.Compose([
    transforms.Resize((64,128))
])
train_data=torchvision.datasets.ImageFolder (root='d:\data\python\cnn_image\Origin_data', transform=trans)
```

```python
# 이미지 크기를 줄이고 다시 확인하면...
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    imshow(data)
    break
```

# 다음 dataset을 사용할때 또 다시 사이즈를 줄여야하므로, 아예 크기를 줄인 이미지를 따로 저장해본다.
# 저장할 폴더는 미리 만들어놔야 저장하면서 에러가 발생안함.

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    if (label == 0):
        data.save('D:\Data\Python\CNN_Image\%d_%d.jpeg'%(num, label))
    elif (label == 1):
        data.save('D:\Data\Python\CNN_Image\%d_%d.jpeg'%(num, label))
    else :
        data.save('D:\Data\Python\CNN_Image\%d_%d.jpeg'%(num, label))
```
